```python
from deepface import DeepFace
import pathlib
import matplotlib.pyplot as plt
import numpy as np
import cv2 
from deepface.commons import distance as dst
```

```python
base_path = pathlib.Path('/media/bao/t7/la_lib_dataset/img')

img1 = str(base_path / "1wBlltV8K9m8vFaKMe2wF6.jpg")
img1_representation = DeepFace.represent(img_path = img1, model_name = 'Facenet512')

img2 = str(base_path / "0T5O55a8aT29CWG_vLgdWD.jpg")
img2_representation = DeepFace.represent(img_path = img2, model_name = 'Facenet512')

img3 = str(base_path / "5Eegciggq6vAXmszqJFueP.jpg")
img3_representation = DeepFace.represent(img_path = img3, model_name = 'Facenet512')
```

```python
def plot_embedding(img_representation):
    img_graph = []
    for _ in range(200):
        img_graph.append(img_representation)
    img_graph = np.array(img_graph)
    im = plt.imshow(img_graph, interpolation='nearest', cmap=plt.cm.ocean)
    plt.colorbar()

def plot_img(img_path):
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
```

```python
plot_img(img1)
```

```python
plot_img(img3)
```
